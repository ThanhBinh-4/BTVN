### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv)

```python
url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv'
df = pd.read_csv(url)
```

### Step 3. Assign it to a variable titanic 

```python
titanic=pd.read_csv(url)
```

### Step 4. Set PassengerId as the index 

```python
df.set_index('PassengerId', inplace=True)
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
sex_counts = df['Sex'].value_counts()
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%')
plt.title('Male/Female Proportion')
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
x = df['Age']
y = df['Fare']
colors = df['Sex'].map({'female':'red', 'male':'blue'})
plt.scatter(x, y, c=colors)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Fare by Age')
plt.show()
```

### Step 7. How many people survived?

```python
df['Survived'].value_counts()[1]
survivor=df['Survived'].sum()
print(survivor)
```

### Step 8. Create a histogram with the Fare payed

```python
plt.hist(df['Fare'], bins=20)
plt.title('Fare Paid in Titanic')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```
