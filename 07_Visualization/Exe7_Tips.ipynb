### Step 1. Import the necessary libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

```python
url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv'
df = pd.read_csv(url)
```

### Step 3. Assign it to a variable called tips

```python
tips=pd.read_csv(url)
```

### Step 4. Delete the Unnamed 0 column

```python
df = df.drop('Unnamed: 0', axis=1)
```

### Step 5. Plot the total_bill column histogram

```python
plt.hist(df['total_bill'], bins=10, alpha=0.5)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Histogram of Total Bill')
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.scatter(df['total_bill'], df['tip'])
plt.xlabel('Total_bill')
plt.ylabel('Tip')
plt.title('The relationship between total_bill and tip')
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['total_bill'], df['tip'], df['size'], s=50, alpha=0.6)

ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
ax.set_zlabel('Size')
ax.set_title('Relationship of Total Bill, Tip, and Size')
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
mean_total_bill = tips.groupby('day')['total_bill'].mean()

x = mean_total_bill.index
y = mean_total_bill.values
plt.bar(x, y)
plt.title('Average Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Average Total Bill')
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
plt.scatter(df['tip'], df['day'], c=df['sex'].map({'Male': 'red', 'Female': 'black'}))

plt.title('Relationship between tip and day, by sex')
plt.xlabel('Tip')
plt.ylabel('Day')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x='day', y='total_bill', hue='time', data=df)
plt.title('Total Bill per Day and Time')
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
dinner_tips = df[df['time'] == 'Dinner']['tip']
lunch_tips = df[df['time'] == 'Lunch']['tip']
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].hist(dinner_tips, bins=15, color='green')
axs[0].set_title('Dinner Tips')

axs[1].hist(lunch_tips, bins=15, color='blue')
axs[1].set_title('Lunch Tips')

plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
male_data = df[df['sex'] == 'Male']
female_data = df[df['sex'] == 'Female']
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].scatter(male_data[male_data['smoker'] == 'Yes']['total_bill'], male_data[male_data['smoker'] == 'Yes']['tip'], color='r', label='Smoker')
axs[0].scatter(male_data[male_data['smoker'] == 'No']['total_bill'], male_data[male_data['smoker'] == 'No']['tip'], color='b', label='Non-Smoker')
axs[0].set_xlabel('Total Bill')
axs[0].set_ylabel('Tip')
axs[0].set_title('Male')
axs[0].legend()

axs[1].scatter(female_data[female_data['smoker'] == 'Yes']['total_bill'], female_data[female_data['smoker'] == 'Yes']['tip'], color='r', label='Smoker')
axs[1].scatter(female_data[female_data['smoker'] == 'No']['total_bill'], female_data[female_data['smoker'] == 'No']['tip'], color='b', label='Non-Smoker')
axs[1].set_xlabel('Total Bill')
axs[1].set_ylabel('Tip')
axs[1].set_title('Female')
axs[1].legend()
plt.show()
```
