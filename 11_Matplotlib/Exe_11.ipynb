```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv'
df = pd.read_csv(url)
#print(df.head(13))
```

```python
#Exercise 1: Read Total profit of all months and show it using a line plot

xpoints = df['month_number'].values
ypoints = df['total_profit'].values
plt.plot(xpoints, ypoints)
plt.title("Company profit per month")
plt.xlabel("Month number")
plt.xticks(df['month_number'])
plt.yticks(np.arange(100000, 600000, 100000))
plt.show()
```

```python
#Exercise 2: Get total profit of all months and show line plot with the following Style properties

xpoints = df['month_number'].values
ypoints = df['total_profit'].values
plt.plot(xpoints, ypoints,color="r",marker='o',mfc='k',ms='8', linestyle='--', linewidth='3.5')
plt.title("Company profit per month")
plt.xlabel("Month number")
plt.legend(loc='lower right',labels=['Profit data of last year'])
plt.xticks(df['month_number'])
plt.yticks(np.arange(100000, 600000, 100000))
plt.show()
```

```python
#Exercise 3: Read all product sales data and show it using a multiline plot

xpoints = df['month_number'].values
y1points = df['facecream'].values
y2points = df['facewash'].values
y3points = df['toothpaste'].values
y4points = df['bathingsoap'].values
y5points = df['shampoo'].values
y6points = df['moisturizer'].values
plt.plot(xpoints, y1points, color='b',marker='o',mfc='b',ms='6',linewidth='3')
plt.plot(xpoints, y2points, color='y',marker='o',mfc='y',ms='6',linewidth='3')
plt.plot(xpoints, y3points, color='g',marker='o',mfc='g',ms='6',linewidth='3')
plt.plot(xpoints, y4points, color='r',marker='o',mfc='r',ms='6',linewidth='3')
plt.plot(xpoints, y5points, color='tab:purple',marker='o',mfc='tab:purple',ms='6',linewidth='3')
plt.plot(xpoints, y6points, color='tab:brown',marker='o',mfc='brown',ms='6',linewidth='3')
plt.legend(labels=['Facecream Sales Data','Facewash Sales Data','Toothpaste Sales Data','Bathingsoap Sales Data','Shampoo Sales Data','Moisturizer Sales Data'])
plt.title('Sales data')
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.xticks(df['month_number'])
plt.yticks([1000, 2000,4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.show
```

```python
#Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot

xpoints = df['month_number'].values
ypoints = df['toothpaste'].values
plt.scatter(xpoints, ypoints, color='c',label='Tooth paste Sales Data')
plt.xticks(df['month_number'])
plt.legend()
plt.grid(axis = 'both', linestyle='--')
plt.title('Toothpaste Sales Data')
plt.xlabel('Month number')
plt.ylabel('Number of units Sold')
plt.xticks(df['month_number'])
plt.show()
```

```python
#Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

month_number = df['month_number'].values
facecream = df['facecream'].values
facewash = df['facewash'].values

plt.bar(month_number, facecream, width=0.3, align='edge', label='Face Cream sales data')
plt.bar(month_number, facewash, width=-0.3, align='edge', label='Face Wash sales data')
plt.grid(axis = 'both', linestyle='--')
plt.xticks(df['month_number'])
plt.title('Facewash and facecream sales data')
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.xticks(df['month_number'])
plt.legend(loc='upper left')
plt.figure(figsize=(10, 20))
plt.show()
```

```python
#Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

month_number = df['month_number'].values
facecream = df['bathingsoap'].values
plt.bar(month_number, facecream, width=0.8)
plt.grid(axis = 'both', linestyle='--')
plt.xticks(df['month_number'])
plt.title('Bathingsoap sales data')
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.xticks(df['month_number'])
plt.figure(figsize=(5,16))
plt.show()
#plt.savefig('C:/my_plot.png')
```

```python
#Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

profits = df['total_profit'].values
plt.hist(profits, bins=10,label='Profit data')
plt.xlim(150000,355000)
plt.title('Monthly Total Profit')
plt.xlabel('Profit range in dollar')
plt.ylabel('Actual profit in dollar')
plt.legend(loc='upper left')
plt.show()
```

```python
#Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

data = {
    'month_number': df['month_number'].values,
    'facecream': df['facecream'].values,
    'facewash': df['facewash'].values,
    'toothpaste': df['toothpaste'].values,
    'bathingsoap': df['bathingsoap'].values,
    'shampoo': df['shampoo'].values,
    'moisturizer': df['moisturizer'].values,
    'total_units': df['total_units'].values,
    'total_profit': df['total_profit'].values,
}
df = pd.DataFrame(data)
total_sales = df.sum()[1:7]

plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%')
plt.legend(labels=['FaceWash','FaceCream','ToothPaste','Bathing Soap','Shampoo','Moisturizer'],loc=(0.8, 0))
plt.title('Total Sales by Product Last Year')
plt.show()
```

```python
#Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

bathing_soap_sales = df['bathingsoap']
facewash_sales = df['facewash']
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 5),sharex=True)

ax1.plot(bathing_soap_sales, color='k', marker='o',mfc='k',ms='6',linewidth='3')
ax1.set_yticks(range(7500, 14500, 2500))
ax1.set_ylim([5500, 15000])
ax1.set_title('Sales data of a Bathing Soap')

ax2.plot(facewash_sales, color='r',marker='o',mfc='r',ms='6',linewidth='3')
ax2.set_yticks(range(1500, 2500, 500))
ax2.set_ylim([1000, 2200])
ax2.set_title('Sales data of a Facewash')

fig.text(0.5, 0.04, 'Month Number', ha='center')
fig.text(0.04, 0.5, 'Sales units in number', va='center', rotation='vertical')
plt.xticks(df['month_number'])

plt.show()

```

```python
#Exercise 10: Read all product sales data and show it using the stack plot

monthList = df ['month_number'].tolist()
faceCremeSalesData = df ['facecream'].tolist()
faceWashSalesData = df ['facewash'].tolist()
toothPasteSalesData = df ['toothpaste'].tolist()
bathingSoapSalesData = df ['bathingsoap'].tolist()
shampooSalesData = df ['shampoo'].tolist()
moisturizerSalesData = df ['moisturizer'].tolist()

plt.figure(figsize=(8,5))
colors = ['m', 'cyan', 'red', 'black', 'green', 'yellow']
plt.stackplot(monthList, faceCremeSalesData, faceWashSalesData, toothPasteSalesData, bathingSoapSalesData, shampooSalesData, moisturizerSalesData, labels=['Face cream', 'Face wash', 'Toothpaste', 'Bathing soap', 'Shampoo', 'Moisturizer'],colors=colors)
#plt.xlabel('Month Number')
#plt.ylabel('Sales unit in number')
plt.title('All product sales data using stack plot')
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.show()
```
